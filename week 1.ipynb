```python
import cv2
capp="D:\\benna\\Notes and imp Questions\\EG TUTORIALS\\WhatsApp Video 2022-01-21 at 12.40.29 PM (1).mp4"
cap=cv2.VideoCapture(capp)
if cap.isOpened()==False:
    print("Error in opening file\n")
while cap.isOpened():
    ret,frame=cap.read()
    if ret==True:
        cv2.imshow('Frame',frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy
img_path="C:\\Users\\anura\\OneDrive\\Desktop\\download.jpeg"
image=cv2.imread(img_path)
plt.subplot(2,2,1)
plt.title('Image')
plt.imshow(image)
alpha=-12
beta=100
image2=cv2.convertScaleAbs(image,alpha=alpha,beta=beta)
cv2.imwrite('download.jpeg',image2)
plt.subplot(2,2,2)
plt.title('contrast and brightness')
plt.imshow(image2)
plt.show()
```
